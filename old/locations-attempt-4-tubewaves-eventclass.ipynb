```python
event_ids = np.arange(100, 1000, 1)

all_aics = []
for id in event_ids:
    e = Event(id=id)
    aics = e.aics[2:]
    all_aics.append(aics)
```

```python
# all_aics
```

```python
len(all_aics[0])
```

```python
fig, ax = plt.subplots()
for aics in all_aics:
    for n, aic in enumerate(aics):
        # aic = aics[1]
        aic = (aic - aic.min())/(aic.max() - aic.min())
        ax.plot(aic, alpha=0.5, color='red')
        diff = np.diff(aic, n=5)
        diff = (diff - diff.max())/(diff.max() - diff.min())
        ax.plot(diff, color='dodgerblue', alpha=0.5)
        raw = Event(id=100).stream[n+2].data
        raw = (raw - raw.min())/(raw.max() - raw.min())
        ax.plot(raw)
    break
```
